```python
#Applying the dbpedia function on these unique books
import requests
import pandas as pd
import time
import json
from pathlib import Path
from titlecase import titlecase


# DBpedia SPARQL endpoint
sparql_endpoint = "http://dbpedia.org/sparql"

# Cache file to store results
cache_file = Path("dbpedia_cache.json")

# Load the cache if it exists
if cache_file.exists():
    with open(cache_file, "r") as f:
        cache = json.load(f)
else:
    cache = {}

def get_dbpedia_abstract_and_genre(title, author):
    """
    Fetch abstract and genre for a book using its title and author.
    """
    # Construct the DBpedia URI by replacing spaces with underscores
    dbpedia_uri = f"http://dbpedia.org/resource/{title.replace(' ', '_')}"

    # Construct the SPARQL query
    query = f"""
    PREFIX dbo: <http://dbpedia.org/ontology/>
    PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>

    SELECT ?abstract ?genre ?authorName WHERE {{
      <{dbpedia_uri}> dbo:abstract ?abstract ;
                      dbo:literaryGenre ?genre ;
                      dbo:author ?author .
      ?author rdfs:label ?authorName .
      FILTER (lang(?abstract) = 'en')
    }}
    """

    # Send the query to DBpedia
    response = requests.get(
        sparql_endpoint,
        params={"query": query, "format": "json"}
    )

    # Parse the response
    if response.status_code == 200:
        results = response.json().get("results", {}).get("bindings", [])
        if results:
            # Filter results to match the author
            for result in results:
                author_name = result.get("authorName", {}).get("value", "").lower()
                if author.lower() in author_name:
                    abstract = result.get("abstract", {}).get("value", "No abstract found")
                    genre = result.get("genre", {}).get("value", "No genre found")
                    return abstract, genre
    return "No abstract found", "No genre found"


# Add new columns for abstract and genre
df_unique_books["DBAbstract"] = ""
df_unique_books["DBGenre"] = ""

# Process each row in the DataFrame
for index, row in df_unique_books.iterrows():
    title = row["Book-Title"]
    author = row["Book-Author"]
    
    # Create a unique key for the cache
    key = f"{title}||{author}"
    
    # Check if the results are already in the cache
    if key in cache:
        abstract, genre = cache[key]
    else:
        # Fetch abstract and genre
        abstract, genre = get_dbpedia_abstract_and_genre(title, author)
        # Store the results in the cache
        cache[key] = (abstract, genre)
        # Add a delay to avoid hitting rate limits (e.g., 1 second delay)
        time.sleep(1)
    
    # Update the DataFrame
    df_unique_books.at[index, "DBAbstract"] = abstract
    df_unique_books.at[index, "DBGenre"] = genre
    
    # Print progress
    print(f"Processed {index + 1}/{len(df)}: {title} by {author}")

# Save the updated DataFrame to a new CSV file (optional)
# df.to_csv("books_with_abstracts_and_genres.csv", index=False)

# Save the cache to a file
with open(cache_file, "w") as f:
    json.dump(cache, f)

# Print the updated DataFrame
print(df_unique_books)
```
