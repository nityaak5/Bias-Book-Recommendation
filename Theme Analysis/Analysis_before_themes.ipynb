```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def calculate_text_length_thresholds(column, multiplier=1.5):
    """Calculate reasonable bounds for text length, ensuring lower bound isn't negative"""
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    
    # Calculate bounds, 0 cuz text length min is 0
    lower_bound = max(0, Q1 - multiplier * IQR) 
    upper_bound = Q3 + multiplier * IQR
    
    return {
        'Q1': Q1,
        'Q3': Q3,
        'IQR': IQR,
        'lower_bound': lower_bound,
        'upper_bound': upper_bound,
        'outliers_low': (column < lower_bound).sum(),
        'outliers_high': (column > upper_bound).sum()
    }

# Calculate thresholds for both columns
words_stats = calculate_text_length_thresholds(df['num_words'])
chars_stats = calculate_text_length_thresholds(df['num_chars'])

# Print results
print("num_words stats:", words_stats)
print("num_chars stats:", chars_stats)

# Filter the dataframe
clean_df = df[
    (df['num_words'] >= words_stats['lower_bound']) & 
    (df['num_words'] <= words_stats['upper_bound']) &
    (df['num_chars'] >= chars_stats['lower_bound']) & 
    (df['num_chars'] <= chars_stats['upper_bound'])
]

print(f"\nOriginal rows: {len(df)}")
print(f"Filtered rows: {len(clean_df)}")
print(f"Rows removed: {len(df) - len(clean_df)}")
```

```python
# Add custom stopwords
custom_stopwords = ["life", "novel", "story", "new", "book", "time", "world", "bestseller", "author", "writer"]
stop_words = set(stopwords.words('english')).union(custom_stopwords)

# Update the vectorizer with custom stopwords and bigrams
vectorizer = TfidfVectorizer(max_features=5000, stop_words=list(stop_words) )
X_tfidf = vectorizer.fit_transform(df['cleaned_description'])

# Retrain LDA
lda = LatentDirichletAllocation(n_components=10, random_state=42)
lda.fit(X_tfidf)
```

```python
def display_topics(model, feature_names, n_top_words=15):
    topics = []
    for topic_idx, topic in enumerate(model.components_):
        top_words = [feature_names[i] for i in topic.argsort()[:-n_top_words - 1:-1]]
        topics.append(top_words)
        print(f"Topic {topic_idx}: {', '.join(top_words)}")
    return topics

# Get feature names (words) from the vectorizer
feature_names = vectorizer.get_feature_names_out()

# Display topics
topics = display_topics(lda, feature_names)
```

```python
def plot_topic_word_distributions(model, feature_names, n_top_words=10):
    n_topics = model.components_.shape[0]
    fig, axes = plt.subplots(n_topics, 1, figsize=(8, 4 * n_topics))
    
    for topic_idx, topic in enumerate(model.components_):
        top_words = [feature_names[i] for i in topic.argsort()[:-n_top_words - 1:-1]]
        weights = topic[topic.argsort()[:-n_top_words - 1:-1]]
        
        ax = axes[topic_idx]
        ax.barh(top_words, weights, color='skyblue')
        ax.set_title(f"Topic {topic_idx}")
        ax.invert_yaxis()  # Invert y-axis to show the highest weight at the top
    
    plt.tight_layout()
    plt.show()
```

```python
topic_distributions = lda.transform(X_tfidf)

# Apply t-SNE
tsne = TSNE(n_components=2, random_state=42)
topic_tsne = tsne.fit_transform(topic_distributions)

# Plot t-SNE
plt.figure(figsize=(10, 8))
plt.scatter(topic_tsne[:, 0], topic_tsne[:, 1], c=topic_distributions.argmax(axis=1), cmap='tab10', s=5)
plt.colorbar()
plt.title('t-SNE visualization of LDA topics')
plt.show()
```

```python
#Trying K-means clustering 

df = df.drop_duplicates(subset=['ISBN'], keep='first')

import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

custom_stopwords = ["life", "novel", "story", "new", "book", "time", "world"]
stop_words = set(stopwords.words('english')).union(custom_stopwords)

# Step 1: Convert text into TF-IDF features
# vectorizer = TfidfVectorizer(max_df=0.8, min_df=2, stop_words=list)
vectorizer = TfidfVectorizer(max_features=5000, stop_words=list(stop_words) )
X = vectorizer.fit_transform(df['cleaned_description'])

# Step 2: Apply K-Means Clustering
num_clusters = 10
kmeans = KMeans(n_clusters=num_clusters, random_state=42, n_init=10)
df['cluster'] = kmeans.fit_predict(X)

# Step 3: Print Top Words in Each Cluster
def get_top_words_per_cluster(n_words=10):
    order_centroids = kmeans.cluster_centers_.argsort()[:, ::-1]
    terms = vectorizer.get_feature_names_out()
    
    for i in range(num_clusters):
        top_words = [terms[ind] for ind in order_centroids[i, :n_words]]
        print(f"Cluster {i}: {', '.join(top_words)}")

get_top_words_per_cluster()

# Step 4: Visualizing Clusters (Optional)
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X.toarray())

plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=df['cluster'], cmap='rainbow', alpha=0.5)
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("K-Means Clusters of Book Descriptions")
plt.show()
```
